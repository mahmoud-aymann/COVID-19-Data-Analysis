## Dataset Contents
  * **Country/Region**: Name of the country or region.
  * **Confirmed**: Total confirmed cases.
  * **Deaths**: Total deaths.
  * **Recovered**: Total recoveries.
  * **Active**: Active cases.
  * **New cases, New deaths, New recovered**: Latest updates.
  * **Deaths/100 Cases**: Case fatality rate percentage.
  * **Recovered/100 Cases**: Recovery rate percentage.
  * **Confirmed last week, 1 week change, 1 week % increase**: Weekly trends.

```python
new_df = pd.concat([num_logged_data, df.select_dtypes(include='O')], axis=1)
new_df
```

```python
### Split X, y 
new_df.dropna(inplace=True)
x = new_df.drop('New deaths', axis=1)
y = new_df['New deaths']
```

```python
#### Split data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
x_train['Country/Region'].nunique()
```

```python
x_train['WHO Region'].nunique()
```

```python
x_train.drop('WHO Region', axis=1, inplace=True )
x_test.drop('WHO Region', axis=1, inplace=True )
```

```python
# !pip install category_encoders
```

```python
from category_encoders import BinaryEncoder 
```

```python
be = BinaryEncoder(cols=['Country/Region'])
```

```python
x_train = be.fit_transform(x_train)
x_test = be.transform(x_test)
```

```python
x_train[x_train['Deaths / 100 Recovered']==np.inf] = x_train[x_train['Deaths / 100 Recovered']==np.inf].replace(np.inf,0)
x_test[x_test['Deaths / 100 Recovered']==np.inf]= x_test[x_test['Deaths / 100 Recovered']==np.inf].replace(np.inf,0)

```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error , r2_score
pred = lr.predict(x_test)
mean_absolute_error(y_test, pred )
```

```python
mean_absolute_error(y_train, lr.predict(x_train))
```

```python
r2_score(y_test, pred)
```
